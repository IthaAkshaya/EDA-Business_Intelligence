```python
import pandas as pd
from google.colab import files

# STEP 1: Upload your CSV file
print("=== Upload File ===")
uploaded = files.upload()
filename = list(uploaded.keys())[0]

# STEP 2: Load dataset
df_raw = pd.read_csv(filename, sep="|", index_col=False)

# Remove empty rows and unnamed columns
df_raw = df_raw.dropna(how='all')
df_raw = df_raw.loc[:, ~df_raw.columns.str.contains('^Unnamed')]
df_raw.columns = df_raw.columns.str.strip()
df_raw.columns = df_raw.columns.str.replace(r"\s+", "", regex=True)
df_raw.columns = df_raw.columns.str.replace(r"\|", "", regex=True)

print("=== TASK 1: RAW DATASET (Before Cleaning) ===")
print(df_raw)

# RAW Data Dictionary
data_dict_raw = pd.DataFrame({
    "Column": df_raw.columns,
    "DataType": [str(df_raw[col].dtype) for col in df_raw.columns],
    "NonNullCount": [df_raw[col].notnull().sum() for col in df_raw.columns],
    "UniqueValues": [df_raw[col].nunique() for col in df_raw.columns]
})
print("\n=== RAW DATA DICTIONARY ===")
print(data_dict_raw)
print("Total rows (raw):", len(df_raw))

# STEP 3: Cleaning
df_clean = df_raw.drop_duplicates()

# Standardize Gender
df_clean['Gender'] = df_clean['Gender'].astype(str).str.strip().str.capitalize()
df_clean['Gender'] = df_clean['Gender'].replace({'M':'Male','F':'Female','Female':'Female','Male':'Male'})

# Convert numeric columns
df_clean['Quantity'] = pd.to_numeric(df_clean['Quantity'], errors='coerce')
df_clean['Price'] = pd.to_numeric(df_clean['Price'], errors='coerce')

# Remove negative prices
df_clean = df_clean[df_clean['Price'] >= 0]

# Parse DOB and PurchaseDate
df_clean['DOB'] = pd.to_datetime(df_clean['DOB'], errors='coerce')
df_clean['PurchaseDate'] = pd.to_datetime(df_clean['PurchaseDate'], errors='coerce')

# Fill missing DOBs
if df_clean['DOB'].notna().any():
    median_dob = df_clean['DOB'].dropna().median()
    df_clean['DOB'] = df_clean['DOB'].fillna(median_dob)
else:
    df_clean['DOB'] = pd.Timestamp("1990-01-01")  # default if all missing

# Fill missing PurchaseDate
if df_clean['PurchaseDate'].notna().any():
    median_date = df_clean['PurchaseDate'].dropna().median()
    df_clean['PurchaseDate'] = df_clean['PurchaseDate'].fillna(median_date)
else:
    df_clean['PurchaseDate'] = pd.Timestamp.today()  # default if all missing

# Calculate Age
df_clean['Age'] = (pd.Timestamp("2024-01-01") - df_clean['DOB']).dt.days // 365

# Fill missing Names
df_clean['Name'] = df_clean['Name'].fillna("Unknown")

print("\n=== TASK 2: CLEANED DATASET (After Cleaning) ===")
print(df_clean)

# CLEANED Data Dictionary
data_dict_clean = pd.DataFrame({
    "Column": df_clean.columns,
    "DataType": [str(df_clean[col].dtype) for col in df_clean.columns],
    "NonNullCount": [df_clean[col].notnull().sum() for col in df_clean.columns],
    "UniqueValues": [df_clean[col].nunique() for col in df_clean.columns]
})
print("\n=== CLEANED DATA DICTIONARY ===")
print(data_dict_clean)
print("Total rows (cleaned):", len(df_clean))

# STEP 4: Export cleaned dataset
df_clean.to_excel("cleaned_sales_data.xlsx", index=False)
print("\nExcel file exported successfully: cleaned_sales_data.xlsx")

# STEP 5: Download cleaned dataset
files.download("cleaned_sales_data.xlsx")
```
