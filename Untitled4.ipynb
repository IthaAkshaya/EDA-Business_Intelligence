```python


# === Step 0: Upload and Load Excel File (Data Tier) ===
from google.colab import files
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

print("=== Upload File ===")
uploaded = files.upload()

print("=== Load Excel File ===")
df = pd.read_excel("cleaned_sales_data (1).xlsx")   # use exact uploaded name
print(df.head())

# === Step 1: Data Cleaning ===
print("=== Data Cleaning Started ===")
df['Gender'] = df['Gender'].astype(str).str.strip().str.capitalize()
df['Gender'] = df['Gender'].replace({'M':'Male','F':'Female','female':'Female'})
df['Age'] = df['Age'].fillna(df['Age'].median())
df['DOB'] = df['DOB'].fillna("Unknown")
df = df.drop_duplicates()
df['PurchaseDate'] = pd.to_datetime(df['PurchaseDate'])
print("=== Cleaned Dataset ===")
print(df.to_string())

# === Step 2: Descriptive Statistics & Univariate Analysis ===
print("=== Descriptive Statistics ===")
print(df.describe())

print("=== Product Frequency ===")
print(df['Product'].value_counts())

print("=== Gender Frequency ===")
print(df['Gender'].value_counts())

# === Step 3: Revenue Calculation ===
df['Revenue'] = df['Quantity'] * df['Price']
print("=== Revenue Column Added ===")
print(df[['TransactionID','Product','Quantity','Price','Revenue']].head(20))

# === Step 4: Basic Visualizations ===
print("=== Revenue by Product ===")
sns.barplot(x="Product", y="Revenue", data=df, estimator=sum)
plt.show()

print("=== Age Distribution ===")
sns.histplot(df['Age'], bins=10, kde=True)
plt.show()

print("=== Revenue by Gender ===")
sns.barplot(x="Gender", y="Revenue", data=df, estimator=sum)
plt.show()

# === Step 5: SQL Queries ===
import sqlite3
print("=== SQL Queries in Colab ===")
conn = sqlite3.connect(':memory:')
df_sql = df.copy()
df_sql['DOB'] = df_sql['DOB'].astype(str)
df_sql['PurchaseDate'] = df_sql['PurchaseDate'].astype(str)
df_sql.to_sql('sales_data', conn, index=False, if_exists='replace')

query = """
SELECT Product, SUM(Revenue) AS TotalRevenue
FROM sales_data
GROUP BY Product
ORDER BY TotalRevenue DESC
LIMIT 5;
"""
result = pd.read_sql(query, conn)
print("=== Top 5 Products by Revenue (Overall) ===")
print(result)

# === Step 6: Top 5 Products by Revenue (Last 6 Months) ===
cutoff_date = df['PurchaseDate'].max() - pd.DateOffset(months=6)
last6 = df[df['PurchaseDate'] >= cutoff_date]
top5_last6 = last6.groupby('Product')['Revenue'].sum().sort_values(ascending=False).head(5)
print("=== Top 5 Products by Revenue (Last 6 Months) ===")
print(top5_last6)

# === Step 7: Monthly User Acquisition Trend ===
df['PurchaseMonth'] = df['PurchaseDate'].dt.to_period('M')
monthly_users = df.groupby('PurchaseMonth')['CustomerID'].nunique()
print("=== Monthly User Acquisition Trend ===")
print(monthly_users)
monthly_users.plot(kind='bar', title="Monthly User Acquisition")
plt.show()

# === Step 8: Multivariate Analysis ===
print("=== Scatter Plot: Age vs Revenue by Product ===")
sns.scatterplot(x="Age", y="Revenue", hue="Product", data=df)
plt.show()

print("=== Scatter Plot: Quantity vs Revenue by Gender ===")
sns.scatterplot(x="Quantity", y="Revenue", hue="Gender", data=df)
plt.show()

# === Step 9: Correlation Analysis ===
print("=== Correlation Heatmap ===")
corr = df[['Quantity','Price','Age','Revenue']].corr()
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f")
plt.show()

print("=== Pair Plot: Age, Quantity, Price, Revenue ===")
sns.pairplot(df[['Age','Quantity','Price','Revenue']], diag_kind="kde")
plt.show()

# === Step 10: KPI Calculations ===
print("=== KPI Calculations ===")

# Total Revenue
total_revenue = df['Revenue'].sum()
print("Total Revenue:", total_revenue)

# Average Age of Customers
avg_age = df['Age'].mean()
print("Average Age of Customers:", round(avg_age, 2))

# Top Product by Revenue (Last 6 Months)
top_product_last6 = last6.groupby('Product')['Revenue'].sum().idxmax()
print("Top Product (Last 6 Months):", top_product_last6)

# Monthly New Users (latest month)
latest_month = monthly_users.index.max()
latest_users = monthly_users.loc[latest_month]
print("Monthly New Users (Latest Month):", latest_users)
```
